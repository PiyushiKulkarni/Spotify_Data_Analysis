```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
df_tracks=pd.read_csv('Sptracks.csv')
df_tracks.tail(20)
```

```python
#check for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
#top 10 least popular songs
sorted_df=df_tracks.sort_values('popularity',ascending = True).head(10)
print(sorted_df)
```

```python
#descriptive statistics of numerical values in the set
df_tracks.describe().transpose()
```

```python
#most popular songs
most_popular=df_tracks.query('popularity > 90.0', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
#changing release date format as index
df_tracks.set_index("release_date",inplace= True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#artist present at 18th row
df_tracks[["artists"]].iloc[18]
```

```python
#df_tracks["duration_ms"]=int(df_tracks["duration_ms"])
```

```python
#duration in ms to sec

df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x:(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
round(df_tracks.duration.head())
```

```python
#visualization of correlation heatmap between variables
corr_df =df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6));
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#taking 0.4% of total data and plotting 2 regression plots
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
#regression plot btween loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data =sample_df, y="loudness", x="energy" ,color="c").set(title="Loudness vs energy correlation plot")

#this shows that there is very high correlation between loudness and correlation
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data =sample_df, y="popularity", x="instrumentalness" ,color="b").set(title="Popularity vs instrumentalness correlation plot")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values(0)
df_tracks.dates=pd.to_datetime(df_tracks.dates,dayfirst=True)
years=df_tracks.dates.dt.year
years
```

```python
#distribution plot - histogram 
sns.displot(years ,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
#duration of songs over the years
total_dr = df_tracks.duration
fig_dims = (18,7)#dimensions
fig, ax= plt.subplots(figsize = fig_dims)#matplotlib subplot func
fig = sns.barplot(x=years, y=total_dr, ax=ax , errwidth= False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
#duration of songs - lineplot
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)#dimensions
fig, ax= plt.subplots(figsize = fig_dims)#matplotlib subplot func
fig = sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
df_genre=pd.read_csv('SpotifyFeatures.csv')
df_genre["genre"]
```

```python
#duration of the songs for different genres/types using barplot color palette
plt.title("Duration of the songs in different genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel("duration in mili seconds")
plt.ylabel("Genres")
```

```python
#top 5 genres by popularity
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity",ascending=False).head(15)#why 15 bcoz there are some features which are appearing twice or more eg.children's music
sns.barplot(y='genre',x='popularity',data=famous).set(title= "Top 5 genres by popularity")
```
